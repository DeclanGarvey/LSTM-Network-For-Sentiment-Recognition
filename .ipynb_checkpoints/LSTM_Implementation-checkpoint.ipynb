```python
# A fuction that takes a string input and "cleans" the string, removing special characters, hyperlinks, extra spaces, etc.

def CleanText(text):
  text = re.sub(r'@\w+|#\w+', '', text) #Removing @mentions
  test = re.sub(r'-', ' ', text)
  text = re.sub('#', '', text) # Removing '#' hash tag
  text = re.sub(r'http\S+', '', text)
  text = re.sub('RT[\s]+', '', text) # Removing RT
  text = re.sub('\w+:\/\/\S+', '', text) # Removing hyperlink
  text = text.translate(str.maketrans('', '', string.punctuation))
  #remove punctuation
  text = re.sub('[^a-zA-Z]', ' ', text)
  text = re.sub(r'\d+', '', text)
  text = text.strip()
  text = text.lower()
  return text
```

```python
# Read the data in  as pandas dataframe cleaning the data, dropping duplicates, and empty entries

InputData = pd.read_csv('./tweet_emotions.csv')
InputData.content = InputData.content.apply(CleanText)
InputData = InputData.drop_duplicates(subset='content', keep="first")
InputData = InputData.dropna()
InputData.sample(10)
```

```python
# hate and anger are merged to be one emotion as they are not sufficiently large datasets and they should appear moderately similarly in text
InputData.loc[(InputData.sentiment=='anger'), 'sentiment'] = 'hate'

# The emotions boredom, empty, and enthusiasm are removed as there is not a sufficiently amount of training which would lead bias in training
InputData = InputData[(InputData['sentiment']!='anger') & (InputData['sentiment']!='boredom') & (InputData['sentiment']!='empty') & (InputData['sentiment']!='enthusiasm')]
Emotions = InputData.sentiment.unique()
Emotions
```

```python
# The sentences are tokenized, this refers to the sentence being split into a list of words

InputData['content'] = InputData['content'].apply(lambda x: nltk.word_tokenize(x))
InputData = InputData[['sentiment','content']]
InputData.sample(5)
```

```python
TrainingData = InputData.groupby('sentiment').apply(lambda x: x.sample(math.floor(len(x)*0.8)).reset_index()).set_index('index')
TestingData = InputData.drop(TrainingData.index) 
```

```python
TrainingData.groupby('sentiment').size()
```

```python
# As can be seen of the training datasets are a lot smaller than others 
# Since similar datasets for each out should used to not inhibit training of your model either all datasets to compare with the smallest
# or all datasets be increase to compare with the largest

# The approach taken in this increase the size of the dataset to match largest or reduce the size by sampling 
# This is done through augmentation 


augsynonym = naw.SynonymAug(aug_src='wordnet',aug_max=3) # This augmenter will replace words in the sentance with there commony used synonyms to a max of 3 words
augswap = naw.RandomWordAug(action='swap', aug_max=3) # This augmenter will randomly swap words within the sentance

DataSetSize = 5000 #TrainingData.groupby('sentiment').size().max()
for emotion in Emotions:
    TempDf = TrainingData[TrainingData['sentiment']==emotion]
    ExtraLinesNeeded = DataSetSize - len(TempDf)

    if(ExtraLinesNeeded>0):
      LineMultiplication = math.ceil(ExtraLinesNeeded/len(TempDf))
      newlines = []
      for m in tqdm(range(math.ceil((LineMultiplication+1)/2))):
        TempDf = TempDf.sample(len(TempDf))
        for message in TempDf[:math.floor(len(TempDf)/2)].content.values:
          newlines.append(augsynonym.augment(message, n=1))
          if(len(newlines)>=ExtraLinesNeeded):
            break
        for message in TempDf[math.floor(len(TempDf)/2):].content.values:
          newlines.append(augswap.augment(message))
          if(len(newlines)>=ExtraLinesNeeded):
            break
        Addon = pd.DataFrame(columns = TempDf.columns)
        Addon['content'] = newlines
        Addon['sentiment'] = emotion
        TrainingData = pd.concat([TrainingData, Addon]).reset_index(drop=True)

TrainingData = TrainingData.groupby("sentiment").sample(DataSetSize)
TrainingData = TrainingData.sample(len(TrainingData))
```

```python
TrainingData.groupby('sentiment').size()
```

```python
# A dicitonary object that will be map each emotion to a torch tensor that will be suitable comparison with the output of a pytorch model is produced

Emotions = TrainingData.sentiment.unique() 
EmotionDic = {}

for i in range(len(Emotions)):
  temp = torch.zeros(len(Emotions))
  temp[i] = 1
  EmotionDic[Emotions[i]] = temp

EmotionDic
```

```python
# A dictionary that will map each words to an integer that will be a suitable input of embedding layer 

l=[]
for i in pd.concat([TrainingData,TestingData]).content:
  l.extend(i)
c = Counter(l)
sorted_by_freq_tuples = sorted(c.items(), key=lambda x: x[1], reverse=True)
ordered_dict = OrderedDict(sorted_by_freq_tuples)
vocab_to_int={w:i+1 for i,w in enumerate(ordered_dict)}
```

```python
#Make Input data matrix 

SequenceLength = 32
def GetInputsAndOutputs(df):
  X_data = pd.DataFrame()
  X_data['content'] = df['content'].apply(lambda x: torch.tensor([*map(vocab_to_int.get,x)])) # map all words in the input data to their corresponding integer

  #FixLength of all inputs to SequneceLength by either cropping to decrease or padding with '0' to increase length
  X_data['content'] = X_data.apply(lambda x: x[:SequenceLength],axis=1) # Crop inputs
  X_data['temp'] = [random.random() for i in range(len(X_data))] # randomly generate a float
  X_data = X_data.apply(lambda x: F.pad(x.content, pad = ( math.floor((SequenceLength-len(x.content))*x.temp) ,math.ceil((1-x.temp)*(SequenceLength-len(x.content)))),value = 0),axis=1) # randomly 0 to start and end to increase to Sequence length
  X_data = torch.stack(X_data.tolist())

  #Make output Matrix
  Y_data = df['sentiment'].map(EmotionDic) # map each emotion to its corresponding torch vector
  Y_data = torch.stack(Y_data.tolist())
  return [X_data.to(device), Y_data.to(device)]


X_temp, Y_temp = GetInputsAndOutputs(TrainingData)
X_test, Y_test = GetInputsAndOutputs(TestingData)

#Randomly split data into training, and validation datasets
X_train, X_val, Y_train, Y_val,  = train_test_split(X_temp,Y_temp,test_size=0.33)
```

```python
#Create dataLoaders with a defined batch size 

batch_size = 256

train_ds = TensorDataset(X_train, Y_train)
test_ds = TensorDataset(X_test, Y_test)
val_ds = TensorDataset(X_val, Y_val)

train_dl = DataLoader(train_ds, shuffle=True, batch_size=batch_size, drop_last=True)
val_dl = DataLoader(val_ds, shuffle=True, batch_size=batch_size, drop_last=True)
test_dl = DataLoader(test_ds, shuffle=True, batch_size=batch_size, drop_last=True)
```

**Embedding layer:**

A layer consisting of a simple look up table which is generated at initialisation which maps a predefined that map set of integers to a vectors of defined length.

**Recurrent Neural Networks:**
Typically for natural language processing in humans you don't simple start from the middle of a sentance ranomly read words, or forget previous when you read new word. What is in fact done is you start from left to right and acquire data as you work your way through sentance.This was a major short coming of classical dense layer neural networks which don't acquire leading to the invention of Necurrent Neural Networks (RNN).

As can be seen from the figure below the network takes the outnput of the adjecent neuron as an extra seperably and training input which emphasis the effect of order and of the input data. Unfortunately, the problem is not selved here, these networks are not good a picking up on long term relationships is commonly present in sentence, i.e. a word at the begining of a sentance can influence the meaning of a word at the end of the sentance. Calling for the ned of further developement networks.

**Long Short-Term Memory(LSTM) Layer:**

LSTM layer help to remedy as they are great at picking up on long term relationships. Although they have higher a complexity then that of a simple RNN they still operator on the same principle.

Further reading on how LSTM layers intrinisically work can be found in the reference.

```python
class SentimentAnlysisModel(torch.nn.Module) :
    def __init__(self, vocab_size, embedding_dim, hidden_dim,number_of_layers, dropout,OutputDim) :
        super().__init__()
        self.hidden_dim = hidden_dim
        self.emb_dim=embedding_dim
        self.no_layers = number_of_layers

        self.embedding = nn.Embedding(vocab_size, embedding_dim,scale_grad_by_freq=True)
        self.lstm = nn.LSTM(input_size=embedding_dim,
                            hidden_size=hidden_dim,
                            num_layers=number_of_layers, 
                            batch_first=True, 
                            dropout=dropout,
                            bidirectional=True
                            )
        self.final = nn.Sequential(nn.Linear(2*hidden_dim,OutputDim))

    def forward(self, x):
        batch_size= x.shape[0]
        len = x.shape[1]
        h0 = torch.zeros(2*self.no_layers, batch_size, self.hidden_dim).to(device)
        c0 = torch.zeros(2*self.no_layers, batch_size, self.hidden_dim).to(device)
        x = x.long()
        
        out = self.embedding(x)
        out, hidden = self.lstm(out, (h0,c0))
        out = out[:, -1]
        out = self.final(out)
        return out
```

Tuning the learning rate is a delicate process as with too high a learning rate it is possible you will over shot the minimum, whilst on the other hand, too low a learning rate your model may get stuck in a local minimum and never reach the global minimum.

In this section, a search for the optimimum learning rate is performed by initially starting with a very low learning rate which is gradually increased with each epochs and the loss recorded.

```python
epochs = 1

# The model initial model is selected
model = SentimentAnlysisModel(len(vocab_to_int)+1, 64, 32, 2, 0.3, len(Emotions))
model = model.to(device)
model.train()

criterion = nn.CrossEntropyLoss()
losses=[]

LearningRates = np.logspace(-8, -0.5, 30)

for lr in LearningRates:
  optimizer = torch.optim.Adam(model.parameters(),lr=lr)
  for e in range(epochs):
      TrainLossSum=0
      num_batches = len(train_dl)
      for batch_idx, batch in tqdm(enumerate(train_dl),total=num_batches):
          
          input = batch[0].to(device)
          target = batch[1].to(device)
          with torch.set_grad_enabled(True):
              out = model(input)
              loss = criterion(out, target)
              loss.backward() 
              optimizer.step()
              optimizer.zero_grad()
  losses.append(loss.item()/input.shape[0])
```

The graph shows the typically loss curve associated with this sort of training, with a global be minimum being visible around 5e-2, which indicative of the optimum learning rate.

```python
plt.figure()
plt.plot(LearningRates, losses)
plt.xscale('log')
plt.yscale('log')
plt.xlabel('Learning rate')
plt.ylabel('Cross Entropy Loss')
plt.show()
```

With a optimum leanring rate found, the model is then initialised, the model is trained for large number of epoch. 

```python
epochs = 20
trainlosses=[]
vallosses = []

model = SentimentAnlysisModel(len(vocab_to_int)+1, 32, 32,3, 0.3, len(Emotions))
model = model.to(device)
criterion = nn.CrossEntropyLoss()
optimizer = torch.optim.Adam(model.parameters(),lr=5e-2)

for e in range(epochs):
    model.train()
    TrainLossSum=0
    num_batches = len(train_dl)
    print(f'Epoch: {e}/{epochs}',end ="   ")
    for batch_idx, batch in  tqdm(enumerate(train_dl), total=num_batches):
      input = batch[0].to(device)
      target = batch[1].to(device)
      with torch.set_grad_enabled(True):
          out = model(input)
          loss = criterion(out, target)
          loss.backward() 
          optimizer.step()
          optimizer.zero_grad()
          TrainLossSum+=loss.item()/input.shape[0]
            
    TrainLossSum/=num_batches
    print(f'Train Loss: {TrainLossSum}',end ="   ")
    trainlosses.append(TrainLossSum)
    
    model.eval()
    ValLossSum=0
    for batch_idx, batch in enumerate(val_dl):
        input = batch[0].to(device)
        target = batch[1].to(device)
        with torch.set_grad_enabled(False):
            #print(input.shape)
            out = model(input)
            loss = criterion(out, target)
            ValLossSum+=loss.item()/input.shape[0]
    ValLossSum/=len(val_dl)
    print(f'Val Loss: {ValLossSum}')
    vallosses.append(ValLossSum)
```

It can be seen from the graph below that over learning on the training is present in but diverengence between the Training and Validiation curve stablises with increasing epoch.

```python
import matplotlib.pyplot as plt
plt.figure()
plt.plot(trainlosses, label='Train')
plt.plot(vallosses, label='Validation')
plt.xlabel("Epoch")
plt.ylabel("Cross-Entropy Loss")
plt.legend()
plt.show()
```

A confusion matrix is formed to indicate where exactly the misclassification errors are occuring. The confusion matrix indicates a group distibution with no anomaly distriubtions indicative of a major error in the model.

The testing accuracy was given to be $\approx 22 $ %, which isn't a bad accuracy as random guessing would result in an accuracy of $\approx 100/9 \approx 11 $ %, indicating the model is in fact training on different features of the dataset. I believe given more data and higher number of training epochs the model would be achieve a much higher accuracy. 

It is also worth noting that the dataset cannot be considered clean as human error is not accounted due to either missclassification or misinterpretation of texts emotion.

```python
y_pred = []
y_true = []
model.eval()
for batch_idx, batch in enumerate(test_dl):
    input = batch[0].to(device)
    target = batch[1].to(device)
    with torch.set_grad_enabled(False):
        out = model(input)
        #print(input.shape)
        #output = net(inputs) # Feed Network

        out = (torch.max(torch.exp(out), 1)[1]).data.cpu().numpy()
        y_pred.extend(out) # Save Prediction
        
        target = torch.max(target, 1)[1].data.cpu().numpy()
        y_true.extend(target) # Save Truth


# Build confusion matrix
cf_matrix = confusion_matrix(y_true, y_pred)
cf_matrix = cf_matrix/np.expand_dims(cf_matrix.sum(axis=1),1)
plt.figure(figsize = (20,15))

for i in range(len(Emotions)):
    for j in range(len(Emotions)):
        text = plt.text(j, i, np.round(cf_matrix[i, j], 2),ha="center", va="center", color="w",size=17)

plt.imshow(cf_matrix)
plt.ylabel('True Class', size=25)
plt.xlabel('Predicted Class', size=25)
plt.xticks(np.arange(Emotions.shape[0]), Emotions,rotation='vertical', size=25)
plt.yticks(np.arange(Emotions.shape[0]), Emotions, size=25)
plt.show()

print(f'Testing Accuracy = {(np.expand_dims(y_true,1)==np.expand_dims(y_pred,1)).sum()/len(y_true)}')
```

```python
# A function that takes an sentance input as a string as an input, then preforming all neccessary preproccessing gives the predicted emotion based on the pretraining model
def ReadyInput(X):
  X = CleanText(X)
  X = word_tokenize(X)
  print(X)
  X = torch.tensor([*map(lambda x: vocab_to_int.get(x, 0),X)])
  X = X[:SequenceLength]
  r = random.random()
  X = F.pad(X, pad = ( math.floor((SequenceLength-len(X))*r) , math.ceil((1-r)*(SequenceLength-len(X)))),value = 0)
  X = X.view(1,X.shape[0])
  return Emotions[model(X.to(device)).argmax()]
```
